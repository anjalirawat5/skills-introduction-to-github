#  Dynamic Typing

Dynamic typing in Python means that variables do not have a fixed data type. Instead, the type of a variable is determined at runtime based on the value assigned to it. This allows for greater flexibility, as you can assign different types of values to the same variable without explicitly declaring its type.



````python
# We are trying to illustrate what we call the dynamic typing
## we can use jar to store to rice or water
## we dn't use same container that is used to rice, we don not use same conatiner to store milk

a = 10
print(type(a))

# it stores in integer data type

a = "india"
print(type(a))

# it stores in string data type

n = 10
print(type(n))

n=n/2
print(type(n))
````

# *Use of Quotes supported by Python Language

* In Python, quotes are used to define string literals. Python supports three types of quotes for strings:

* Single Quotes (' ')
* Double Quotes (" ")
* Triple Quotes (''' ''' or """ """)

# 1. Single Quotes (' ')

* Used for simple strings.
* Cannot contain another single quote unless escaped.

2. Double Quotes (" ")
* Works the same as single quotes but allows single quotes inside without escaping.

3. Triple Quotes (''' ''' or """ """)
* Used for multi-line strings.
* Preserves formatting (line breaks, indentation).
* Commonly used for docstrings (function or class documentation).
* It is used for adding multi-line comments.



````python
x = 'This is a string'
y = "This is also a string"
z = '''first line
second line
third line'''

print(x)
print(y)
print(z)
````
